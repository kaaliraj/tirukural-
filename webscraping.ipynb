```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}

startpoint =1
periods =[]
count = 0
for startpoint in range(1,1331):
    r = requests.get("https://www.ytamizh.com/thirukural/kural-"+str(startpoint), headers=headers)
    c = r.content
    soup = BeautifulSoup(r.content, 'html5lib')
    table = soup.find('div', attrs = {'name':'kural-'+str(startpoint)}) 
    period = [re.sub('\n',' \n ',pt.get_text()) for pt in table.find_all('p')]
    count = count+1
    periods.append(period)
    print (count)
    
pd.DataFrame(periods).to_excel('tirukural.xlsx')
pd.DataFrame(periods).to_csv('tirukural.csv')
```
